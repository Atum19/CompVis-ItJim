```python
def mach_templates(img, fin_img, **kwargs):
    tm_bgr = kwargs['tm_bgr']
    norm_method = kwargs['norm_method']
    thresh_coef = kwargs['thresh_coef']
    colour = kwargs['colour']
    if kwargs['rotate_flag']:
        tm_bgr = cv2.rotate(tm_bgr, kwargs['rotate_method'])
    
    matching = cv2.matchTemplate(img, tm_bgr, norm_method)
    matching = cv2.normalize(matching, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

    thresh = cv2.threshold(matching, np.max(matching)*thresh_coef, 255, cv2.THRESH_BINARY)[1]   # for 1  # rotating also needed  ROTATE_90_COUNTERCLOCKWISE

    # Finding contours of the maximal coincidence points
    conts, hier = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Drawing rectangles to highlight the entire template image on the initial image for each possible case
    for cnt in conts:
        x, y, w, h = cv2.boundingRect(cnt)
        cv2.rectangle(fin_img, (x+w//2, y+h//2), (x+w//2+tm_bgr.shape[1], y+h//2+tm_bgr.shape[0]), colour, 6)
        cv2.putText(fin_img, kwargs['label'], (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, colour, 1)
        
    return fin_img
```
